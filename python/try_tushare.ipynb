```python
data=data.sort_index(ascending=True)
close_prices=[price for price in data['close'].values]
data['ma3'] = np.round(pd.Series.rolling(data['close'], 3).mean(), 2)
data['ma5'] = np.round(pd.Series.rolling(data['close'], 5).mean(), 2)
data[['ma5']].plot(figsize=(12, 6))
#data['ma5'] = np.round(pd.Series.rolling(data['close'], 5).mean(), 2)
#data[['ma3','ma5']].plot(figsize=(12, 7))
print(data[['ma3','ma5']])
```

```python
data['open'].values
```

```python
# 策略： 以当日收盘价计算，如果ma3_price > ma5_price，则次日以开盘价买入；如果ma3_price < ma5_price，则次日以开盘价卖出。
# 以2016-09-28日起始，从2016-10-11开始计算，每次只交易100股，按5元一次计算手续费。
signals=[]
isBuy=False
dateBuy=''
priceBuy=0
isSell=False
dateSell=''
priceSell=0
profitTotal=0
feeTotal=0
assets=0
i=0
for index in data['ma5'].index:
    signal = 1 # 买入
    if data['ma3'][index] > data['ma5'][index]:
        signal = 1
    else:
        signal = 0
    
    signals.append(signal)
    #print("%s, open %.2f, close %.2f, %d" % (index, data['open'][index], data['close'][index], signal))
    if i > 3:
        # if isBuy:
        #     print('buy')
        #     
        # if isSell:
        #     prict('sell')
            
        if signals[i] > signals[i-1]:
            isBuy = True
            feeTotal += 5
            dateBuy = index
            priceBuy = data['close'][index]
            assets = priceBuy * 100
            #print(index, 'buy price %.2f' % priceBuy)
        
        if signals[i] < signals[i-1]:
            isSell = True
            feeTotal += 5
            dateSell = index
            priceSell = data['close'][index]
            profit = priceSell*100 - assets - 10
            profitTotal += profit
            assets = 0
            print("buy price %.2f on %s, sell price %.2f on %s, profit %d, total profit %d" % 
                  (priceBuy, dateBuy, priceSell, dateSell ,profit, profitTotal))

    i+=1
```
